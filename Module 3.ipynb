```python
def calc(A,B,C):
    pl = int(A+B+C)
    sum =0
    i = 2
    while(pl!=0):
        
        rem = pl%10
        sum+=rem*i
        pl//=10
        i+=1
    if(sum%11 == 10):
        return '0'
    return str(sum%11)
```

```python
def ChangeHex(n):
    if (n<=1):
        return (str(n))
    else:
        ans = ChangeHex( n // 16 )
        x =(n%16)
        s = str(x)
        if (x < 10):
            return (ans+s)  
        if (x == 10):
            return (ans+"A")
        if (x == 11):
            return (ans+"B")
        if (x == 12):
            return (ans+"C")
        if (x == 13):
            return (ans+"D")
        if (x == 14):
            return (ans+"E")
        if (x == 15):
            return (ans+"F")
```

```python
def fix(mapping, pl, descriptions):
    
    pl = str(pl)
    key = 0
    A = pl[0:2]
    B = pl[2:4]
    C = pl[5:7]
    D = pl[7]
    
    for dsc in descriptions:
        hex = ChangeHex(key)
        if(len(hex)<2):
            hex = '00'+hex
        elif(len(hex)<3):
            hex = '0'+hex
            
        allotedPl = A+B+hex+D#calc(A,B,key)
        mapping[allotedPl] = [dsc]
        key+=1
        

#         flag = False
#         while(not(flag) and key<1000):
            
#             allotedPl = int(A+B+str(key)+calc(A,B,str(key)))
#             if allotedPl not in mapping:
#                 mapping[allotedPl] = [dsc]
#                 flag = True
#             else:
#                 key+=1
#         if(not flag):
#             print("Too many data items.")
    
    
    mapping.pop(pl,None)  
    
    
    return mapping   
```

```python
mapping = {}
for row in data:
    if row[0] not in mapping:
        mapping[row[0]] = [row[1]]
    else:
        mapping[row[0]].append(row[1])
    
pl = sorted(mapping.keys())

for num in pl:
    val = mapping[num]
    if(len(val)>1):
        print('\033[30;1;46m',num,':',mapping[num])
    else:    
        print('\033[30;1;m',num,':',mapping[num])


    
```

```python
for num in pl:
    val = mapping[num]
    if(len(val)>1):
        mapping = fix(mapping,num,val)
        
print('\n\n\n###########\n\n')
for key in mapping:
    print(key,':', mapping[key])
```
